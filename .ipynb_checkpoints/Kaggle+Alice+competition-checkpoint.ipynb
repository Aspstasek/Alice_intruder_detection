```python
y = train_df["target"]
```

```python
train_df.drop("target", axis = 1, inplace = True)
```

```python
X = train_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 17)
```

```python
tree = DecisionTreeClassifier(random_state=7)
```

```python
train_test_split?
```

```python
params = {"max_depth":range(2,12),"max_features":[0.25,0.5,0.75,1]}
tree_grid = GridSearchCV(tree,params,n_jobs=-1)
```

```python
tree_grid.fit(X_train,y_train)
```

```python
test_pred = tree_grid.best_estimator_.predict(X_test)
```

```python
accuracy_score(test_pred,y_test)
```
